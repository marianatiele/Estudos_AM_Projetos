```python
#Bibliotecas
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import  skfuzzy as cms
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carregar a base de dados
dados = datasets.load_iris()
```

```python
type(dados)
```

```python
#Criando o cluester cms.cmeans(), matriz transposta de data.T =  dados.data.T, c=3 -> número de cluster

l = cms.cmeans(data = dados.data.T, c= 3,m=2, error=0.0005, maxiter=100, init=None)
```

```python
#Mostrando o cluster
l
```

```python
#Porcentagem  de um registro do clusters, que está na posição 1
previsao_porcentagem = l[1]
previsao_porcentagem
```

```python
#Visualizando a porcentagem de cada cluster
for i in range(150):
  print(previsao_porcentagem[0][i], previsao_porcentagem[1][i], previsao_porcentagem[2][i])
```

```python
#Matriz de confusão
previsoes = previsao_porcentagem.argmax(axis=0)
resultado = confusion_matrix(dados.target, previsoes)
```
